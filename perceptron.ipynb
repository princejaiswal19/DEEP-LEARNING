```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

X, y = make_classification(
    n_samples=100,  # Corrected from n_sample
    n_features=2,  # Corrected from n_fatures
    n_informative=1,
    n_redundant=0,
    n_classes=2,
    n_clusters_per_class=1,
    random_state=41,
    class_sep=10  # No need to specify hypercube=False in most cases
)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")
plt.show()
```
